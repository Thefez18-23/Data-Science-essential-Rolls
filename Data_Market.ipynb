The first str.replace('£', '') removes the pound currency symbol (£).
	•	The second str.replace('K', '000') replaces ‘K’ (as in “K” for thousand) with ‘000’.


```python
data['Salary'] = data['Salary'].str.replace('£', '').str.replace('K', '000')
```

```python
data.head(10)
```

```python
data['Salary'] = data['Salary'].str.replace(r'\(employer est.\)', '', regex=True)
data['Salary'] = data['Salary'].str.replace(r'\(Glassdoor est.\)', '', regex=True)
```

```python
import re 
```

```python
def convert_to_yearly(salary):
    if 'Per Hour' in str(salary):
        # Use re.findall to extract the numeric value from the salary string
        hourly_rate = re.findall(r'\d+\.?\d*', salary)
        if hourly_rate:
            # Convert the first found number to a float and multiply by 2000 to convert to yearly salary
            return float(hourly_rate[0]) * 2000
    return salary  # Return the original salary if no hourly rate was found
data['Salary'] = data['Salary'].apply(convert_to_yearly)
```

```python
data.head(10)
```

```python
data['Salary'] = data['Salary'].astype(str)
```

```python
data[['salary_min', 'salary_max']] = data['Salary'].str.split('-', expand=True)
```

```python
data.head(10)
```

```python
data['salary_min'] = data['salary_min'].str.strip()
data['salary_max'] = data['salary_max'].str.strip()
```

```python
# Optionally, you can convert salary_min and salary_max to numeric if they should be numbers
data['salary_min'] = pd.to_numeric(data['salary_min'], errors='coerce')
data['salary_max'] = pd.to_numeric(data['salary_max'], errors='coerce')
```

```python
data['Location'] = data['Location'].astype(str)
```

```python
data[['city', 'country']] = data['Location'].str.split(',',n=1, expand=True)
```

```python
data.head(10)
```

```python
skills_split = data['Skills'].str.get_dummies(sep=', ')
```

```python
skills_split
```

```python
data_with_skills = pd.concat([data, skills_split], axis=1)
data_with_skills.head()
```

```python
from wordcloud import WordCloud
```

data_with_skills.iloc[:, data_with_skills.columns.get_loc('AI'):]:
	•	The iloc[:,] selects all rows.
	•	data_with_skills.columns.get_loc('AI') finds the position (index) of the column named ‘AI’.
	•	The : after get_loc('AI') ensures that all columns from ‘AI’ onwards are selected.
	•	.sum(): Sums the values across all rows for the selected columns. By default, this sums the values vertically (column-wise sum).

```python
# Correcting the code to calculate the sum of columns starting from the 'AI' column
skill_frequencies = data_with_skills.iloc[:, data_with_skills.columns.get_loc('AI'):].sum()
```

```python
skill_frequencies
```

```python
%matplotlib inline
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(skill_frequencies)

# Display the word cloud
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Hide axes
plt.title('Most Popular Skills Word Cloud', fontsize=16)
plt.show()
```
